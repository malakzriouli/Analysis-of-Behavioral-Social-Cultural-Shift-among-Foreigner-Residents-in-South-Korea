**Text Analysis of question : Could you specify why?(stress)**

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

text_columns = [
    'Could you specify why?(stress)'
]

# Combine all text columns into a single text field per row
text_data = df[text_columns].fillna('')
combined_text = text_data.apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

# Clean the text
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]
    return ' '.join(tokens)

cleaned_text = combined_text.apply(clean_text)

# Vectorize single words
vectorizer = CountVectorizer(max_features=30)
X = vectorizer.fit_transform(cleaned_text)
word_freq = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Print top 15 frequent words
print("Top 15 frequent words:")
print(word_freq.sum().sort_values(ascending=False).head(5))

# Vectorize bigrams (2-word combinations)
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2), max_features=10)
X_bigrams = bigram_vectorizer.fit_transform(cleaned_text)
bigrams = pd.DataFrame(X_bigrams.toarray(), columns=bigram_vectorizer.get_feature_names_out())

print("\nTop 10 frequent bigrams:")
print(bigrams.sum().sort_values(ascending=False).head(10))

# Create word cloud
text_string = ' '.join(cleaned_text.tolist())
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(text_string)

plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud")
plt.show()

# Topic modeling using LDA
lda = LatentDirichletAllocation(n_components=4, random_state=42)
lda.fit(X)

print("\nLDA Topic Modeling - Top Words Per Topic:")
for idx, topic in enumerate(lda.components_):
    print(f"Topic {idx + 1}: ", [vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
```

# **Sentiment analysis**

```python
from transformers import pipeline
import pandas as pd

# Load your survey data (make sure to adjust the file and column name)
#df = pd.read_csv("/content/CSS survey 91.csv")  # Replace with your actual CSV file name
responses = df["Could you mention why?(life in korea long run)"]  # Replace with your column name

# Initialize sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# Run sentiment analysis on each response
sentiments = []
for response in responses:
    if isinstance(response, str) and response.strip():  # Skip empty responses
        result = sentiment_analyzer(response[:512])[0]  # Truncate to 512 tokens max
        sentiments.append({"text": response, "label": result['label'], "score": result['score']})
    else:
        sentiments.append({"text": "", "label": "NEUTRAL", "score": 0})

# Convert results to DataFrame
sentiment_df = pd.DataFrame(sentiments)

# Merge with original data if needed
df = pd.concat([df, sentiment_df[["label", "score"]]], axis=1)

# Save results to new CSV
df.to_csv("survey_sentiment_results.csv", index=False)

print("Sentiment analysis complete. Results saved to 'survey_sentiment_results.csv'.")
```

**Korean Level VS Social Activity (Chi-square test)**




```python
import pandas as pd
from scipy.stats import chi2_contingency

# 1. Load and clean relevant columns
df_clean = df[['Korean level', 'Since moving to South Korea, are you socially more active?']].copy()

# 2. Clean up values (remove whitespace, unify categories)
df_clean['Korean level'] = df_clean['Korean level'].str.strip().str.lower()
df_clean['socially_active'] = df_clean['Since moving to South Korea, are you socially more active?'].str.strip().str.lower()

# 3. Optional: map Korean levels to a logical order (if needed for plotting or scoring)
level_order = {
    'dont know korean': 0,
    'beginner': 1,
    'intermediate': 2,
    'advanced': 3
}
df_clean['korean_level_num'] = df_clean['Korean level'].map(level_order)

# 4. Create a contingency table
contingency_table = pd.crosstab(df_clean['Korean level'], df_clean['socially_active'])

# 5. Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Test Results:")
print(f"Chi2 value: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

# 6. Optional: Interpret
if p < 0.05:
    print("\n✅ There is a statistically significant relationship between Korean level and being socially active.")
else:
    print("\n❌ No statistically significant relationship between Korean level and being socially active.")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is in a dataframe called 'df'
# If reading from CSV: df = pd.read_csv("your_file.csv")

# Step 1: Create ordered categorical for Korean level
df['Korean_level_ordered'] = pd.Categorical(df['Korean level'],
                                           categories=["Dont know Korean", "Beginner", "Intermediate", "Advanced"],
                                           ordered=True)

# Step 2: Clean up the adaptation column name for easier reference
#adaptation_col = 'Which of the following best describes how you have adapted to Korean culture?\n(Please read the definitions carefully before choosing.)'
adaptation_col ='Adaptation Type'
# Step 3: Create contingency table
contingency_table = pd.crosstab(df['Korean_level_ordered'], df[adaptation_col])
print("Contingency Table:")
print(contingency_table)
print()

# Step 4: Perform Chi-square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("Chi-square Test Results:")
print(f"Chi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Result: Statistically significant association (p < 0.05)")
else:
    print("Result: No statistically significant association (p >= 0.05)")
print()

# Step 5: Calculate Cramér's V (effect size)
n = contingency_table.sum().sum()  # Total sample size
cramers_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))

print(f"Cramér's V: {cramers_v:.4f}")

# Effect size interpretation for Cramér's V
if cramers_v < 0.1:
    effect_size = "negligible"
elif cramers_v < 0.3:
    effect_size = "small"
elif cramers_v < 0.5:
    effect_size = "medium"
else:
    effect_size = "large"

print(f"Effect size: {effect_size}")
print()

# Step 6: Display expected frequencies (for chi-square assumptions)
print("Expected Frequencies:")
expected_df = pd.DataFrame(expected,
                          index=contingency_table.index,
                          columns=contingency_table.columns)
print(expected_df.round(2))
print()

# Check chi-square assumptions
cells_below_5 = (expected < 5).sum().sum()
total_cells = expected.size
print(f"Chi-square assumption check:")
print(f"Cells with expected frequency < 5: {cells_below_5} out of {total_cells}")
if cells_below_5 / total_cells > 0.2:
    print("Warning: More than 20% of cells have expected frequency < 5")
    print("Consider using Fisher's exact test or combining categories")
print()

# Step 7: Calculate percentages within each Korean level
print("Percentage Distribution within each Korean Level:")
percentage_table = pd.crosstab(df['Korean_level_ordered'], df[adaptation_col], normalize='index') * 100
print(percentage_table.round(1))
print()

# Step 8: Visualizations
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8))

# Stacked bar chart
contingency_table.plot(kind='bar', stacked=True, ax=ax2, colormap='Set3')
ax2.set_title('Adaptation Types by Korean Level (Counts)', fontsize=14)
ax2.set_xlabel('Korean Level', fontsize=12)
ax2.set_ylabel('Count', fontsize=12)
ax2.legend(title='Adaptation Type', bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)

# Heatmap of percentages
sns.heatmap(percentage_table, annot=True, fmt='.1f', cmap='Blues', ax=ax1)
ax1.set_title('Adaptation Types by Korean Level (Percentages)', fontsize=14)
ax1.set_xlabel('Adaptation Type', fontsize=12)
ax1.set_ylabel('Korean Level', fontsize=12)

plt.tight_layout()
plt.show()

# Step 9: Summary of findings
print("=" * 60)
print("SUMMARY OF FINDINGS:")
print("=" * 60)
print(f"• Chi-square statistic: {chi2:.4f}")
print(f"• P-value: {p_value:.4f}")
print(f"• Cramér's V: {cramers_v:.4f} ({effect_size} effect)")
if p_value < 0.05:
    print("• There IS a statistically significant association between Korean level and adaptation type")
else:
    print("• There is NO statistically significant association between Korean level and adaptation type")

# NOTES:
# 1. Chi-square test checks if adaptation type and Korean level are independent
# 2. Cramér's V measures the strength of association (0 = no association, 1 = perfect association)
# 3. The contingency table shows the actual counts for each combination
# 4. The percentage table shows what % of each Korean level group chose each adaptation type
```
